```
import pandas as pd

# Load the dataset
df = pd.read_csv("data.csv")
print(df)
```

```
# Display the first 10 rows
print(df.head(10))

```

```
# Inspect the schema
print(df.info())
```

```
# Data cleaning
df = df.dropna()  # Drop rows with null values
df = df[df['ARR_DELAY'] > 0]  # Filter out rows where ARR_DELAY is less than or equal to zero
```

```
# Aggregation and Summary Statistics
grouped_data = df.groupby('CARRIER')['ARR_DELAY'].mean()
print(grouped_data)
```

```
min_delay = df['ARR_DELAY'].min()
max_delay = df['ARR_DELAY'].max()
mean_delay = df['ARR_DELAY'].mean()
print(f"Minimum Delay: {min_delay}")
print(f"Maximum Delay: {max_delay}")
print(f"Mean Delay: {mean_delay}")
```

```
import matplotlib.pyplot as plt
```

```
# Bar chart
df['CARRIER'].value_counts().plot(kind='bar')
plt.title('Total Flights per Carrier')
plt.show()
```

```
# Histogram
df['ARR_DELAY'].plot(kind='hist', bins=50)
plt.title('Distribution of Arrival Delays')
plt.xlabel('Arrival Delay (minutes)')
plt.ylabel('Frequency')
plt.show()
```
